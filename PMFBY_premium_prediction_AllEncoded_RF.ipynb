# Creating an additional column 'Policy Term'.
#### (period during which the policyholder is covered by the insurance policy).

```python
def calculate_duration(df):
    """
    Calculates the duration in days between 'startDate' and 'endDate' 
    and adds a new column 'Policy Term' to the DataFrame.
    """
    # Convert 'startDate' and 'endDate' to datetime objects, handling errors
    df['startDate'] = pd.to_datetime(df['startDate'], format='%Y-%m-%d', errors='coerce')
    df['endDate'] = pd.to_datetime(df['endDate'], format='%Y-%m-%d %H:%M:%S', errors='coerce')


    # Calculate duration in days
    df['Policy Term (Days)'] = (df['endDate'] - df['startDate']).dt.days

    return df

# Calculate and add duration column
pmfby_duplicate_removed = calculate_duration(pmfby_duplicate_removed.copy())

# View the Policy Term column
pmfby_duplicate_removed[['Policy Term (Days)']]
```

### Outlier handling should be based on nature of data and the impact of outliers on the selected model

#### Outliers as Errors:
If outliers are due to data entry errors or inconsistencies, they should be corrected or removed. For example, if a farmer's premium is mistakenly recorded as ₹1 instead of ₹1000.
#### Outliers as Genuine Values:
In some cases, outliers might represent actual high-value crops or policies. For example, a large insurance claim or premium for a commercial farm.

# Frequency encoding: 'categoryName', 'State', 'insuranceCompanyName'

```python
import pandas as pd

def frequency_encode(df, cols):
  """
  Performs frequency encoding on specified columns in a DataFrame.

  Args:
    df: The DataFrame to encode.
    cols: A list of column names to be encoded.

  Returns:
    The DataFrame with new columns containing frequency encoded values.
  """

  for col in cols:
    freq_map = df[col].value_counts().to_dict()
    df[col + '_freq'] = df[col].map(freq_map)

  return df

df = pmfby_duplicate_removed

# Columns to encode
cols_to_encode = ['categoryName', 'State', 'insuranceCompanyName']

# Perform frequency encoding
frequency_encoded_df = frequency_encode(df, cols_to_encode)

display(frequency_encoded_df) 
```

```python
from sklearn.preprocessing import OneHotEncoder
# Specify the columns to encode
columns_to_encode = ['Year','isPreviousSeasonYearInSubsidy', 'firstGoiSubsidy']
columns_to_drop= ['Year','categoryName', 'State', 'insuranceCompanyName','isPreviousSeasonYearInSubsidy', 'firstGoiSubsidy'] #List of all encoded columns


# Initialize OneHotEncoder
encoder = OneHotEncoder(sparse_output=False, drop='first') #Ensures the output is a dense NumPy array instead of a sparse matrix

# Fit and Transform Only the Specified Columns
encoded_array = encoder.fit_transform(frequency_encoded_df[columns_to_encode])

# Get Feature Names for Encoded Columns
encoded_feature_names = encoder.get_feature_names_out(input_features=columns_to_encode)

# Create a DataFrame from the Encoded Array
oneHot_encoded_df = pd.DataFrame(encoded_array, columns=encoded_feature_names)

# Drop the Original Encoded Columns and Concatenate the new Encoded Columns
frequency_encoded_df = frequency_encoded_df.drop(columns=columns_to_drop).reset_index(drop=True)
encoded_df = pd.concat([frequency_encoded_df, oneHot_encoded_df], axis=1)

print("Final DataFrame:")
encoded_df
```

# Hyperparameter Tuning for SVR

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

# Define parameter grid for RandomForestRegressor
param_grid_rf = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}

# Perform Grid Search for RandomForestRegressor
grid_search_rf = GridSearchCV(estimator=RandomForestRegressor(),
                              param_grid=param_grid_rf,
                              cv=5, scoring='neg_mean_squared_error', verbose=2, n_jobs=-1)

# grid_search_rf.fit(X_train, y_train)
grid_search_rf.fit(X_train, y_train.to_numpy())

# Best hyperparameters and performance for RandomForestRegressor
print("Best Hyperparameters for RandomForestRegressor:", grid_search_rf.best_params_)
best_rf_model = grid_search_rf.best_estimator_
y_pred_rf_tuned = best_rf_model.predict(X_test)
```
